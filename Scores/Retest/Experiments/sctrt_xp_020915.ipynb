# Scores NYU TRT experiment

The purpose of this experiment is to do an initial exploration of the scores maps of one or two individuals across scanning sessions. The main questions are:
1. <div style='color:green'>Starting with any arbitrary network, how variable is the shape and magnitude of the scores map for an individual across retest sessions</div>
2. <div style='color:blue'>Is there a noticeable difference in the variability between the session on the same day (1 and 2) and between those on different days (1/2 and 3)</div>
3. <div style='color:orange'>Comparing two subjects qualitatively, is the variability between subjects larger than the variability between sessions</div>
4. <div style='color:orange'>Looking at an individual subject, are there networks for which the variability is larger or smaller? How does this generalize to other subjects</div>
5. <div style='color:orange'>How does scores compare to other metrics like dual regression and seed maps</div>

## Setup

```python
# Get the imports handled
import os
import glob
import numpy as np
import nibabel as nib
import brainbox as bb
```

```python
# Set the paths for the files that we are working with
scale = 7
metric = 'stability_maps'
in_path = '/data1/scores/retest/out/sc{0:0>2}'.format(scale)
template = '_fmri_sub{0}_session{1}_rest_stability_maps.nii.gz'
sub2 = '22894'
```

## @ 1 Individual network and subject across runs

```python
# Set the current subject names and network numbers
network = 5
# Adjust the network id to zero based indexing so we have a humanly readable variable
net_id = network - 1
sub1 = '05676'
```

### Get the data and plot it

```python
# Set the paths up for the three different sessions
sp_run1 = os.path.join(in_path, metric, template.format(sub1, 1))
sp_run2 = os.path.join(in_path, metric, template.format(sub1, 2))
sp_run3 = os.path.join(in_path, metric, template.format(sub1, 3))

# Load the images
img_1 = nib.load(sp_run1)
img_2 = nib.load(sp_run2)
img_3 = nib.load(sp_run3)

# Get the data from the image
data_1 = img_1.get_data()[..., net_id]
data_2 = img_2.get_data()[..., net_id]
data_3 = img_3.get_data()[..., net_id]

# Get the montages of the network
mont_1 = bb.visu.make_montage(data_1)
mont_2 = bb.visu.make_montage(data_2)
mont_3 = bb.visu.make_montage(data_3)
```

```python
# Make a plot with the three figures side by side
f = plt.figure(figsize=[20, 6])
ax1 = f.add_subplot(131)
ax1.imshow(mont_1, vmin=0, vmax=1)
ax1.set_title('Session 1')

ax2 = f.add_subplot(132)
ax2.imshow(mont_2, vmin=0, vmax=1)
ax2.set_title('Session 2')

ax3 = f.add_subplot(133)
im = ax3.imshow(mont_3, vmin=0, vmax=1)
ax3.set_title('Session 3')

f.suptitle('Subject {} @ network {} ({})'.format(sub1, network, metric))
```

However, there is a higher spatial similarity between the 1st and the 2nd session, especially with respect to laterality. It's not really clear from these images. Therefore I will take a difference map.

## @2 Variability between sessions

```python
# Get an average and take both the distance from the average as well as difference between sessions
stack = np.concatenate((data_1[..., None], data_2[..., None], data_3[..., None]),axis=3)
avg = np.mean(stack, 3)
# Get the distances from the average
da1 = avg - data_1
da2 = avg - data_2
da3 = avg - data_3
# Get the distances between the sessions
d12 = data_1 - data_2
d13 = data_1 - data_3
d23 = data_2 - data_3
# Montage all of these
va1 = bb.visu.make_montage(da1)
va2 = bb.visu.make_montage(da2)
va3 = bb.visu.make_montage(da3)
v12 = bb.visu.make_montage(d12)
v13 = bb.visu.make_montage(d13)
v23 = bb.visu.make_montage(d23)
```

```python
# Show them
f = plt.figure(figsize=[16, 10])
xa1 = f.add_subplot(231)
xa1.imshow(va1, vmin=-1, vmax=1, cmap=cm.RdBu)
xa1.set_title('AVG - 1')

xa2 = f.add_subplot(232)
xa2.imshow(va2, vmin=-1, vmax=1, cmap=cm.RdBu)
xa2.set_title('AVG - 2')

xa3 = f.add_subplot(233)
xa3.imshow(va3, vmin=-1, vmax=1, cmap=cm.RdBu)
xa3.set_title('AVG - 3')

x12 = f.add_subplot(234)
x12.imshow(v12, vmin=-1, vmax=1, cmap=cm.RdBu)
x12.set_title('1 - 2')

x13 = f.add_subplot(235)
x13.imshow(v13, vmin=-1, vmax=1, cmap=cm.RdBu)
x13.set_title('1 - 3')

x23 = f.add_subplot(236)
x23.imshow(v23, vmin=-1, vmax=1, cmap=cm.RdBu)
x23.set_title('2 - 3')
```
