# Clean up the subject file

```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
```

```python
# Paths
pheno_file = '/data1/guilimin/simons_vip/pheno/dataset5.csv'
match_pheno = '/data1/guilimin/simons_vip/pheno/dataset_good_16p.csv'
raw_path = '/data1/guilimin/simons_vip/scores/scores_11_07/rmap_part/'
out_pheno = '/data1/guilimin/simons_vip/pheno/dataset_good_16p_session2.csv'
```

```python
pheno = pd.read_csv(pheno_file, delimiter=',')
pheno.rename(columns={'Unnamed: 0':'Subject'}, inplace=True)
```

```python
pheno
```

```python
pheno.session.value_counts()
```

```python
# Drop these guys
drop = list()
for ind, row in pheno.iterrows():
    if row.session == 1:
        drop.append(ind)
        continue
    # File that we look for:
    fname = '{}_rmap_part.nii.gz'.format(row.Subject)
    # Path that we are looking for
    fpath = os.path.join(raw_path, fname)
    # Does the file exist
    if not os.path.isfile(fpath):
        print('{} does not exist, drop em'.format(fname))
        drop.append(ind)
```

```python
use_pheno = pheno.drop(drop)
```

```python
use_pheno.rename(columns={'Subject':''}, inplace=True)
```

```python
use_pheno.to_csv(out_pheno, index=False)
```
